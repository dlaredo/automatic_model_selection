```python
def get_compiled_model(model, optimizer_params=[]):
    """Obtain a keras compiled model"""
    
    #Shared parameters for the models
    optimizer = Adam(lr=0.001, beta_1=0.5)
    
    lossFunction = "categorical_crossentropy"
    metrics = ["accuracy"]
    
    #Create and compile the models
    model.compile(optimizer = optimizer, loss = lossFunction, metrics = metrics)
    
    return model


def create_tunable_model(model_handler, input_shape, data_handler, model_number):
    
    K.clear_session()
    
    model = model_handler(input_shape)
    
    model = get_compiled_model(model, optimizer_params=[])
    
    model = SequenceTunableModelClassification('ModelClass_SN_'+str(model_number), model, lib_type='keras', data_handler=data_handler)
          
    return model
```

```python
evaluations_cv = list()
evaluations_test = list()
cv_all = []
test_all = []
S05 = [S1_05, S2_05, S3_05, S4_05, S5_05]
S03 = [S1_03, S2_03, S3_03, S4_03, S5_03]
S07 = [S1_07, S2_07, S3_07, S4_07, S5_07]
alphas = [S05, S03, S07]

k = 5 
input_shape = 784

dHandler_mnist = MNISTDataHandler()

#Create CV folds
dHandler_mnist.load_data(verbose = 1, unroll=True)
folds = list(KFold(n_splits=k, shuffle=True).split(dHandler_mnist.X_train))

for alpha in alphas:
    
    print('\nalpha', alpha)
    
    for model_handler in alpha:
    
        print('\nmodel', model_handler)
        k = 5
        input_shape = 784
    
        for j, (train_idx, val_idx) in enumerate(folds):

            print('\nExperiment on Fold ', j)
            
            K.clear_session()  #Clear the previous tensorflow graph 

            model = model_handler(input_shape)
            model = get_compiled_model(model, optimizer_params=[])

            X_train_cv = dHandler_mnist.X_train[train_idx]
            y_train_cv = dHandler_mnist.y_train[train_idx]
            X_valid_cv = dHandler_mnist.X_train[val_idx]
            y_valid_cv = dHandler_mnist.y_train[val_idx]
            
            trainable_count = int(np.sum([K.count_params(p) for p in set(model.trainable_weights)]))
            
            model.fit(X_train_cv, y_train_cv, batch_size=512, epochs=100, verbose=1)
            
            evaluation_cv = model.evaluate(X_valid_cv, y_valid_cv)
            evaluation_test = model.evaluate(dHandler_mnist.X_test, dHandler_mnist.y_test)
    
            print(evaluation_cv)
            print(evaluation_test)
            
        cv_all.append(evaluation_cv[1])
        test_all.append(evaluation_test[1])
        print(cv_all)
        print(test_all)
```
